#Reading file#

```python
import pandas as pd
data = pd.read_excel("University_Clustering.xlsx")
data.head()
```

#Slicing the DataSet#

##For rows##

```python
data = data.iloc[:, :] # So this means, [rows, columns]
data.head() #this will print whole data as it is
```

If you want to slice or cut first 3 rows then --> [3 : , : ] 

```python
dataRow1 = data.iloc[3:, :] #this will cut first 3 rows
dataRow1
```

If you want to slice or cut rows exept first 3 rows then --> [ : 3
 , : ] 

```python
dataRow2 = data.iloc[:3, :] 
dataRow2
```

##For Columns##

If you want to slice or cut first 3 columsn then --> [ : , 3 : ] 

```python
dataCol1 = data.iloc[:, 3:] #this will cut first 3 columns
dataCol1
```

If you want to slice or cut rows exept first 3 columns then --> [ : 
 , : 3] 

```python
dataCol1 = data.iloc[:, :3] #this will cut first 3 rows
dataCol1
```

For, range of the Columns

```python
dataRange = data.iloc[:, 2:5]
dataRange
```

##For Deleting Specific Column/s ##

```python
dataDrop1 = data.drop(["State"], axis = 1) #axis = 1 is for column
dataDrop1
```

```python
dataDrop2 = data.drop(["State", "SAT"], axis = 1)
dataDrop2
```
