#K-means clustering on Randomized data#

##Importing Datasets##

```python
import pandas as pd
import numpy as np
import matplotlib.pylab as plt

from sklearn.cluster import	KMeans
# from scipy.spatial.distance import cdist 
```

```python
X = np.random.uniform(0,1,50) # X-axis 50 points 
Y = np.random.uniform(0,1,50) # Y-axis 50 points
```

```python
#df_xy named variable is created and in that with help of pandas, a empty dataFrame is created having Two Columns X and Y
df_xy = pd.DataFrame(columns = ["X","Y"]) # having Columns 'X' and 'Y'.
```

## Normalization of the data##

The data we have, any of it does not have same scale or unit so that we can compare them, So they sould be Normalized and Standardization is done. So Normalization is done here and all the data is made ranging from 0 to 1 at their perspective.

1. Min max Scalling is done here.
2. It makes data scale free or unitless.
3. So the data is transformed between 0 to 1 unitlessly.


```python
                                            # This is the formula or the function for normalizing the data
def norm_func(i):
    x = (i - i.min())	/ (i.max() - i.min()) # This is the formula of the normalization which makes data
                                            # ranges from 0 to 1.
    return (x)                              # that value is returned.
```

```python
sliced_data = Univ.iloc[:, 1:]
```

```python
sliced_data
```

```python
df_norm = norm_func(Univ.iloc[:, 1:])       # Normalized data frame (considering the numerical part of data)
                                            # Univ data is passed to norm_func for normalization
                                            # In Univ dataset, University name is removed or sliced.
                                            # All the data will be assigned to df_norm
```

```python
#So the data is transformed between 0 to 1
df_norm
```

## Scree plot or Elbow curve ##

Here, we will find value of TWSS to find value of K(number of clusters)

1. Plotting will be done of K and TWSS.
2. x axis = K
3. y axis = TWSS


```python
TWSS = []               # Here Total Within Sum of Sqaure list is created.
k = list(range(2, 9))   # Values for k is assigned with list, starting from 2 to 9. It is not
                        # started from 1, because It'll make TWSS highest and that is worst case (i.e. making only one cluster)
```

```python
for i in k:                         # (for every value of i in k):
    kmeans = KMeans(n_clusters = i) # so all values of k from 2 to 9 will be passed to kMeans(n_clusters = i) and that number of clusters are made and whole is assigned to kmeans.
    kmeans.fit(df_norm)              # and then with that(i) number of clusters, df_norm dataset is fitted.
    TWSS.append(kmeans.inertia_)    # Append the TWSS value to TWSS list which is calculated by .intertia_ function
                                    # therefore, for every value of 'kmeans', .intertia_(i.e. TWSS will be calculated)
                                    # Inbuilt Function :- 'kmeans.inertia_' will tell Within SS rather TotalWithin SS(TWSS)  
```

```python
TWSS # These are the TWSS values and on that basis number of clusters will be choosen having large drop and lowest number of TWSS
     # These are the values of TWSS for K = 2 to k = 9.
```

```python
# Scree plot 
plt.plot(k, TWSS, 'ro-'); plt.xlabel("No_of_Clusters"); plt.ylabel("total_within_SS")
                      # So plotting is done and (k, TWSS, 'ro-') means 
                      # (x axis = k, y axis = TWSS, 'ro-' with red color having circle and dash(-))
                      # labels are provided for x Axis K :- Number of Clusters and for y Axis :- TWSS.
```

Here, we can see frok k = 4 to k = 5, there is large drop and TWSS is lower. So k = 5 can be used.
